## Khởi tạo Dataframe

Khai báo dữ liệu theo cột

```python
import pandas as pd

df = pd.DataFrame({           #Dictionary
    "X" : [13, 30, "A"],      #List
    "Y" : [20, 40, "B"],
    "Z" : [25, 50, "C"],
    "T" : [10, 40, "D"]},
    index = [0, 1, 2])
df
```

Load dữ liệu từ file csv (bảng)

```python
stocks = pd.read_csv("/content/drive/MyDrive/Python_ML/Theory/Data/stocks.csv")
stocks
```

## Góm nhóm dữ liệu

Phương thức pivot

```python
stocks.pivot(index = 'Name', columns = 'Sex', values = 'Age')
```

```python
stocks.pivot(index = 'Name', columns = 'Sex', values = ['Age', 'Pclass'])
```

Phương thức pivot_table

```python
import numpy as np
stocks.pivot_table(index = 'Sex', columns = 'Age',  values = ['Fare', 'Pclass'], aggfunc = np.mean)
```

## Nối dữ liệu

Nối dữ liệu theo chiều dọc với concat (axis = 1)

```python
df1 = pd.DataFrame({           #Dictionary
    "X" : [13, 30, "A"],      #List
    "Y" : [20, 40, "B"],
    "Z" : [25, 50, "C"],
    "T" : [10, 40, "D"]},
    index = [1, 2, 3])
df2 = pd.DataFrame({           #Dictionary
    "A" : [1, "D"],      #List
    "B" : [2, "E"],
    "C" : [3, "F"],
    "D" : [4, "G"]},
    index = [1, 2])

df_new = pd.concat([df1, df2], axis = 1)
df_new
```

## Điền giá trị

Điền giá trị khuyết với fillna(value)

```python
df_new.fillna(1)
```

## Trích xuất dữ liệu

Lấy tập con theo dòng

```python
sub_df = df_new[df_new.Y!=20]
sub_df
```

```python
sub_df = df_new[df_new.Y.isin([40, 'B'])]
sub_df
```

Lấy tập con theo cột

```python
colunms = df1[['X', 'Z']]
print(colunms)
colunms = df1['X'] # hoặc df1.X
colunms
```

## Vẽ biểu đồ cơ bản

Hàm plot và scatter

```python
df_ = pd.DataFrame({           #Dictionary
    "X" : [13, 1, 1],      #List
    "Y" : [20, 40, 3]},
    index = [0, 1, 2])
df_.plot.hist()
df_.plot.scatter(x='X', y='Y')
```
