## 1. Array và khởi tạo array

- Numpy array là một cấu trúc các giá trị có cùng kiểu dữ liệu, được đánh chỉ mục bởi các số nguyên không âm

Để tạo array ta truyền một list vào np.array()

```python
a = np.array([1, '2', 3])
print(a)
a = np.array([1,2,3]) #Array 1 chiều
print(a)
```

```python
print("Kiểu dữ liệu của a là:", type(a))
print("Kích thước từng chiều (dimension) của a là: ", a.shape)
print("Truy xuất 3 phần tử đầu tiên: ", a[0], a[1], a[2])
a[0] = 5
print("Array a sau khi đã cập nhật phần tử đầu tiên: ", a)
```

Chỉ mục là mảng Boolean (Boolean array indexing). Cho phép lấy ra phần tử bất kì trong array gốc thỏa mãn tính chất nào đó

```python
a = np.array([[1,2], [3,4], [5,6]])
bool_idx = a>2
print(bool_idx)
```

```python
# Sử dụng để tạo array 1 chiều với các phần tử thỏa mãn điều kiện
print(a[bool_idx])
# Có thể viết gọn
print(a[a>2])
```

Ngoài ra còn một số hàm:
*   `argmax` (lấy chỉ số phần tử lớn nhất)
*   `argmin` (lấy chỉ số phần tử nhỏ nhất)
*   `argsort` (lấy chỉ số các phần tử sau khi đã sắp xếp thứ tự tăng dần)
*   `where` (lấy chỉ số phần tử thỏa mãn điều kiện nào đó)




```python
a = np.array([1, 4, 3, 7, 6, 9, 5])
print("Chỉ số phần tử lớn nhất trong a: ", a.argmax()) #np.argmax(a)
print("Chỉ số phần tử nhỏ nhất trong a: ", a.argmin()) #np.argmin(a)
print("Chỉ số các phần tử sau khi đã sắp xếp thứ tự tăng dần: ", a.argsort()) #np.argsort(a)
print("Chỉ số các phần tử sau khi đã sắp xếp thứ tự giảm dần: ", a.argsort()[::-1])
print("Chỉ số các phần tử thỏa mãn điều kiện a>5: ", np.where(a>5)[0])
print("Chỉ số các phần tử lớn nhất trong mảng a: ", np.where(a>=np.argmax(a))[0])
```

## 4. Array Math

Numpy có thể thực hiện được các hàm toán học vector hóa (vectorized) để tính toán các phần tử của một array. Các hàm được tối ưu hóa và có tốc độ thực thi nhanh hơn vòng `for` gấp nhiều lần

Ví dụ ta có thể thực hiện phép cộng hai array ngẫu nhiên, sử dụng `%%time` để đo thời gian thực thi

```python
a = np.random.random(100000000)
a
```

```python
%%time
print(np.sum(a))
```

```python
%%time
x = 0
for i in a:
  x += i
print(x)
```

Để nhân ma trận, ta sử dụng hàm `dot()` hoặc toán tử @

```python
x = np.array([[1,2], [3,4]])
y = np.array([[5,6], [7,8]])

v = np.array([9,10])
w = np.array([11,12])

print(x.dot(v)) # Nhân ma trận với vector
print(np.dot(x,v))
print(x@v)
print()
print(x.dot(y)) # Nhân ma trận với ma trận
print(np.dot(x,y))
print(x@y)
print()
print(v.dot(w))
print(np.dot(v,w)) # Nhân 2 vector
print(v@w)
#
```

## Broadcasting

Broadcasting tuân thủ các quy tắc sau:

* Nếu array khác số chiều, kích thước array với số chiều ít hơn sẽ chèn thêm 1 phía đầu bên trái

```python
a = np.array([1,2,3]) # shape(3,): 1 chiều sẽ ngầm mở rộng -> 2 chiều shape(1,3)
b = np.array([[4], [5], [6]]) #shape (3,1): 2 chiều
c = a + b # kết quả là shape(3,3) (2 chiều) -> Vận dụng thêm quy tắc 2
print(c)
```

* Nếu shape của 2 array không khớp ở một chiều nào đó, array nào có kích thước chiều bằng 1 sẽ được kéo dài ở đó chiều đó để khớp với array còn lại

```python
a = np.array([[1], [2], [3]]) # shape(3,1) -> shape(3,2)
# --> [1,1]
#     [2,2]
#     [3,3]
b = np.array([[4,5], [6,7], [8,9]]) # shape(3,2)
c = a + b # shape(3,2)
print(c)
```

- Nếu hay array không có chiều nào bằng nhau và không có chiều nào bằng 1 thì báo lỗi

```python
a = np.array([[1], [2], [3], [4]])
b = np.array([[4,5], [6,7], [8,9]])
# c = a + b -> lỗi
```

## 1. Plotting

Hàm quan trọng nhất của `matplotlib` là `plot`, cho phép vẽ dữ liệu 2D dưới dạng đường thẳng

```python
# Khởi tạo x và y là các tọa độ điểm trên đường cong hình sine
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)

# Vẽ các điểm này với matplotlib
plt.plot(x,y)

# Hiển thị kết quả
plt.show()
```

Ta có thể vẽ nhiều đường thẳng cùng lúc, thêm các tiêu đề, chứ thích và ghi chú trên từng trục dữ liệu

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

# Vẽ các điểm với matplotlib
plt.plot(x, y_sin, label = "sine")
plt.plot(x, y_cos, label = "cosine")
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.title("Sine and Cosine")
plt.legend() # Sử dụng tham số label ở trên để chú thích
plt.show() # Hiển thị kết quả
```

## 2. Subplot

Nếu muốn vẽ nhiều hình khác nhau cùng một lúc, ta dùng `subplot`

```python
# Khởi tạo x và y là các tọa độ điểm trên đường cong hình sine và cosine
x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Tạo lưới subplot với chiều cao là 2, chiều rộng là 1
# Sau đó kích hoạt ô đầu tiên để về
# 2 hàng, 1 cột --> thực hiện vẽ trên ô đầu tiên (ô số 1)
plt.subplot(2, 1, 1)

# Vẽ lên ô đầu tiên
plt.plot(x, y_sin)
plt.title("Sine")

# Kích hoạt ô thứ 2, sau đó vẽ lên ô này
plt.subplot(2, 1, 2)
# 2 hàng, 1 cột, sau đó vẽ lên ô thứ 2
plt.plot(x, y_cos)
plt.title("Cosine")

# Điều chỉnh khoảng cách giữa các ô
plt.subplots_adjust(hspace=0.5)

# Hiển thị kết quả
plt.show()
```

## 3. Chart types

```python
t = np.arange(0, 2, 0.05)
s = np.sin(2*np.pi*t)

# Scatter
plt.subplot(221)
plt.scatter(t, s, marker = ">")
plt.title("Scatter")

# Bar chart
plt.subplot(222)
x = np.arange(5)
y = np.random.rand(5)
plt.bar(x, y)
plt.title("Bar chart")

# Histogram
plt.subplot(223)
plt.hist(s)
plt.title("Histogram")

#Pie chart
plt.subplot(224)
labels = ["Other", "China", "USA", "Việt Nam"]
sizes = [15, 20, 25, 40]
plt.pie(sizes, labels=labels)
plt.title("Pie")

plt.subplots_adjust(hspace=0.5)
plt.show()
```
