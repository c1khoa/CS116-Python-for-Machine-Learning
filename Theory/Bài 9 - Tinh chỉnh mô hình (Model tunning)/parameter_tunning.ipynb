## Tại sao cần phải tinh chỉnh tham số

### Tham số trong máy học

Tham số mô hình (parameter): là các biến số mô hình học từ dữ liệu train
- Ví dụ: trọng số, bias của một Neural Network

Siêu tham số (hyper-parameter): là cấu hình của mô hình trước khi train
- Được thiết lập thủ công thông qua quá trình tinh chỉnh tham số
- Ví dụ: số lớp, số neuron của một lớp trong Neural Network
image.png

### Tại sao cần tinh chỉnh tham số

Hiệu suất mô hình: tinh chỉnh siêu tham số giúp mô hình đạt được độ chính xác cao hơn, hiệu quả hơn

Overfitting và Underfitting: việc chọn siêu tham số không phù hợp có thể dẫn đến overfitting hoặc underfitting

Tài nguyên tính toán: siêu tham số ảnh hưởng đến thời gian và tài nguyên tính toán cần thiết để huấn luyện mô hình

Thích ứng với dữ liệu: tinh chỉnh siêu tham số giúp mô hình được điều chính phù hợp nhất với đặc trưng riêng của từng dữ liệu

## Phương pháp Grid Search

Ý tưởng: "vét cạn" các tổ hợp tham số có khả năng xảy ra, sau đó tiến hành thử và chọn tổ hợp cho kết quả tốt nhất

Ưu điểm:
- Thuật toán đơn giản
- Tìm kiếm toàn diện không gian tham số

Khuyết điểm
- Đánh đồng vai trò các siêu tham số
- Không gian tham số lớn khiến chi phí tính toán lớn
- Không hiệu quả do không kế thừa kết quả tìm kiếm trước đó
image.png

## Phương pháp Random Search

Ý tưởng: chọn  ngẫu nhiên các tổ hợp tham số, thay vì lấy mẫu đều, sau đó thử lần lượt và chọn tổ hợp cho mô hình tốt nhất

Ưu điểm:
- Đơn giản, dễ cài đặt
- Hiệu quả với không gian tìm kiếm lớn

Khuyết điểm:
- Nếu không gian tìm kiếm lớn, khó tìm được tối ưu toàn cục
- Các lần thử độc lập, không kế thừa thông tin các lần thử trước
- Kết quả không nhất quán do yếu tố ngẫu nhiên

image.png

## Phương pháp Bayesian Optimization

Ý tưởng: chiến lược tuần tự để tìm kiếm tôi ưu toàn cục của các mô hình dạng black-box

Ưu điểm:
- Hiệu quat, có kế thừa thông tin các lần thử
- Giảm bớt số lần thử nghiệm
- Phù hợp với không gian tham số lớn

Khuyết điểm
- Phức tạp hơn so với Grid Search, Random Search
- Việc chọn hàm accquisition ảnh hưởng lớn đến hiệu quả của phương pháp
