# Học có giám sát - Mô hình phân loại (Classification)

## Mô hình Logistic Regression

Cho hai tập điểm "xanh" và "cam" được biểu diễn với y = 1 hoặc y = 0

Hàm mô hình dự đoán: $
\hat{y} = \sigma\left( \hat{\beta} x \right)
$

Hàm độ lỗi dự đoán: $$ BCE = -\Sigma^n_{i=1}y_ilog\hat{y} + (1-y_i)log(1-\hat{y_i})$$

Ưu điểm:
- Đơn giản, dễ cài đặt
- Có thể mở rộng sang bài toán phân loại nhiều lớp

Khuyết điểm:
- Không giải quyết được dữ liệu phức tạp
- Dễ bị ảnh hưỡng bởi dữ liệu nhiễu

## Mô hình với dữ liệu có quan hệ phi tuyến

KNNClassifier: là một mô hình phân loại dựa trên nhãn của K đặc trưng có khoảng cách gần nhất -> Lazy learning -> Tốn bộ nhớ

### Multi-Layer Perceptron (MLP)

Multi-Layer Perceptron (MLPClassifier): sử dụng thêm các lớp ẩn (hidden layer) để học các đặc trưng trung gian, giúp việc ra quyết định dễ hơn

### Cây quyết định - Decision Tree

Decision Tree: là một cấu trúc phân cấp không tham số, trong đó
- Mỗi nút đại diện cho một thuộc tính (hay đặc trưng)
- Mỗi nhánh từ nút đó tương ứng với một số khả năng của thuộc tính đó
- Mỗi nút lá (nút cuối cùng) đưa ra quyết định phân loại cuối cùng dựa trên các thuộc tính đã duyệt trước đó

image.png

Dựa trên độ đo Information Gain để quyết định chọn đặc trưng nào để phân loại cho nút tiếp theo

Entropy thể hiện mức độ đa dạng các loại đối tượng trong tập S:

$$ Entropy(S) = -Σ^c_{i=1}p_ilogp_i$$

$$ InforGain = Entropy(Parrent) - E[Entropy(Children)] $$

Ưu điểm:
- Tính giải thích mô hình cao: nhờ khả năng trực quan hóa với cấu trúc cây
- Có thể làm việc trên dữ liệu dạng số lẫn phân loại
- Phi tham số: không đưa ra giả định về sự phân bố của các biến và mối quan hệ giữa các đặc trưng với output
- Lựa chọn đặc trưng: ưu tiên các đặc trưng nhiều thông tin để phân thành nhánh mới

Khuyết điểm:
- Overfitting: dễ bị overfitting với dữ liệu nếu cây không được cắt tỉa (pruning) hoặc cây quá sâu (mô hình quá phức tạp)
- Không ổn định: chỉ cần thay đổi dữ liệu một phần nhỏ, cây có thể thay đổi hoàn toàn -> dùng kĩ thuật Ensemble với Random Forest
- Khó tối ưu: việc tìm cây tối ưu tốn nhiều chi phí tính toán, trong khi dùng heuristic thì không đảm bảo được cây tối ưu
- Bias: có xu hướng bị ảnh hưởng bởi các đặc trưng có nhiều giá trị (biến phân loại), hoặc phạm vi lớn hơn (biến số), vì sẽ tạo nhiều nhánh hơn
