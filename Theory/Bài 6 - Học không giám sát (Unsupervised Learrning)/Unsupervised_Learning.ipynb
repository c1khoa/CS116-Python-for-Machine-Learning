# Học không giám sát - Unsupervised Learning

## Giới thiệu học không giám sát

### Giới thiệu

Học không giám sát (unsupervised learning) là một nhánh của ML, có nhiệm vụ học phân bố của dữ liệu, từ đó có thể biểu diễn dữ liệu hiệu quả hơn.

Dữ liệu cho thuật toán học không giám sát không cần gán nhãn (label)
- Chỉ cần dữ liệu đầu vào x
- Không cần nhãn đầu ra tương ứng

Một số chủ đề chính:
- Gom nhóm dữ liệu - Clustering
- Giảm chiều dữ liệu - Dimension reduction

## Các mô hình gom nhóm - Clustering

Gom nhóm (clustering) là bài toán gom nhóm các đối tượng theo từng cum sao cho các đội tượng trong cùng một cụm có sự tương đồng với nhau hơn so với những đối tượng thuộc các nhóm khác

Một số thuật toán gom nhóm dữ liệu:
- K-Means (gom nhóm khi biết trước số lượng cụm)
- DBSCAN (không biết trước số lượng cụm, gom cụm dựa trên tiêu chí mật độ)

### Thuật toán gom nhóm - KMeans

Ý tưởng: khởi tạo ngẫu nhiên K tâm cụm, sau đó gán các kiểm dữ liệu trọng tâm gần nhất. Quá trình này được lặp đi lặp lại cho đến khi các trọng tâm không thay đổi nữa

Ưu điểm:
- Đơn giản, dễ cài đặt
- Hiệu quả với dữ liệu lớn (> 1M mẫu xài AKMeans)

Khuyết điểm:
- Cần biết trước số lượng cụm K
- Dễ bị rơi vào cực tiểu cục bộ (phụ thuộc vào khởi tạo K cụm)
- Phụ thuộc vào tâm cụm khởi tạo
- Không hoạt động tốt với dữ liệu có phân bố phức tạp, không phải dạng hình cầu

Thuật toán gom nhóm - DBSCAN

Ý tưởng: phân cụm dựa trên mật độ, gom các điểm gần nhay (có khoảng cách nhỏ hơn *e*) và có mật độ cao (số điểm tối thiểu trong cụm là *minPts*)

Các điểm nằm trong các cụm mật độ thấp được gán là nhiễu (noise)
image.png

Ưu điểm:
- Không cần biết trước số cụm
- Hiệu quả với dữ liệu mật độ cao

Khuyết điểm:
- Không hiệu quả khi dữ liệu có mật độ biến động
- Phải chọn tham số *e* và *minPts*

## Các mô hình giảm chiều dữ liệu

Giảm chiều dữ liệu: là quá trình chuyển đổi dữ liệu từ không gian đa chiều sang không gian ít chiều sao cho biễu diễn không gian ít chiều vẫn giữ được một số tính chất có ý nghĩa của dữ liệu gốc

Một số bài toán giảm chiều dữ liệu:
- Phân tích thành phần chính - PCA
- Nhúng t - SNE

Mỗi phương pháp có những đặc điểm và ứng dụng riêng, tùy thuộc vào bộ dữ liệu và mục tiêu cụ thể

### Thuật toán Principal Component Analysis - PCA

image.png

PCA tìm ra một không gian con tuyến mới mà dữ liệu được biểu diễn một cách hiệu quả nhất

Trong không gian mới này, các chiều được chọn sao cho tối đa hóa phương sai của dữ liệu
image.png

### tSNE (t-Distributed Stochastic Neightbor Embedding)

t-SNE: là kỹ thuật giảm chiều phi tuyến tính, để trực quan hóa dữ liệu đa chiều trong không gian có số chiều thấp hơn (thường là 2D hoặc 3D)

Ý tưởng: tạo ra phân phối xác suất tương tự trong không gian có số chiều thấp hơn
image.png

### So sánh PCA và tSNE

PCA:
- Ưu điểm:
  - Nhanh, hiệu quả với dữ liệu do thực hiện phép biến đổi tuyển tính
  - Giữ lại các thành phần chính với phương sai lớn
- Khuyết điểm:
  - Khả năng nắm bắt cấu trúc phi tuyến kém
  - Dễ bị ảnh hưởng bởi nhiễu (outlier)
  
tSNE:
- Ưu điểm:
  - Có khả năng bắt cấu trúc phi tuyến
  - Thường sử dụng để trực quan hóa dữ liệu
- Khuyết điểm:
  - Không ổn định do yếu tố ngẫu nhiên
  - Độ phức tạp cao
  - Khó giải thích hơn so với PCA
