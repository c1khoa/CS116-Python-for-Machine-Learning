## Tại sao cần đánh giá mô hình

Xác thực tính hiệu quả

- Mô hình thực sự đoán chính xác hay chỉ nhớ dữ liệu train?

Sự tin cậy: mô hình của mình có ổn định và đáng tin cậy khi triển khai thực tế không?
Định hướng cải thiện mô hình: các độ đo đánh giá giúp ta biết nên cải tiến mô hình theo hướng nào?

### Phân biệt bài toán học máy

Hai loại bài toán chính:

- Hồi quy - Regression
- Phân lớp - Classification

Cách phân biệt "đơn giản" dựa trên tính chất của biến kết quả (output)

- Hồi quy: đầu ra là giá tị liên tục
- Phân lớp: đầu ra là giá trị rời rạc

Câu hỏi: Bài toán "dự đoán tuổi" là Hồi quy hay Phân lớp?

Cách phân biệt rõ ràng hơn:

- Hồi quy: 2 giá trị output bất kì có thể thực hiện tất cả so sánh >, <, =, !=
- Phân lớp: 2 giá trị output bất kì chỉ có thể thực hiện so sánh = hoặc !=

-> Đáp án: Bài toán dự đoán tuổi là bài toán hồi quy

-> Tương tự, bài toán đoán đối tượng "chó", "mèo" là bài toán phân loại

### Tại sao cần Phân biệt đúng bài toán

Để chọn hàm mục tiêu huấn luyện phù hợp -> Hàm loss đúng việc huấn luyện nhanh hơn

Để chọn lựa độ đo đánh giá phù hợp -> Đánh giá khác quan hơn

Tại sao không sử dụng log loss làm độ đo đánh giá? -> Khó hiểu

-> Sử dụng Laccard index
- Precision = TP/(TP+FP)
- Recall = TP/(TP+FN)
- F1_score = 2(Pr x Re)/(Pr + Rec)

Nhận xét
- Các độ đo đánh giá, hàm độ lỗi trên là cách tính trung bình, thực tế cho thể tính thiên lệch

### So sánh

Test là một phần của train:
- Ưu điểm: đơn giản, nhanh
- Khuyết điểm: không khách quan, tính tổng quát không được đảm bảo

Tran-test split:
- Ưu điểm: đơn giản, dễ thực hiện, it tốn kém chi phí tính toán hơn Cross Validation, phù hợp với tập dữ liệu lớn
- Khuyết điểm: Tốn thêm thời gian, không khách quan bằng Cross Validation

K-Fold Cross Validation
- Ưu điểm: đánh giá toàn diện về tính tổng quát của mô hình, phù hợp với tập dữ liệu ít, huấn luyện nhanh
- Khuyết điểm: chi phí tính toán rất cao
