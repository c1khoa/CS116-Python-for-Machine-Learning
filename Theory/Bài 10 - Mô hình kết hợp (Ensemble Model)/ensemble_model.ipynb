## Ti sao cn c Ensemble Model

### Gii thiu

Mc tiu my hc: xy dng m hnh c tnh tng qut ha cao t d liu

C hai cch chnh  ci thin tnh tng qut ha
- Ci thin hiu sut m hnh my hc
- Kt hp nhiu m hnh v t hp kt qu d on -> Ensemble Learrning

image.png

### Ti sao Ensemble Learning hiu qu

Vn  gim variance:
- S dng nhiu m hnh c th trung bnh gi tr d on gn vi gi tr thc t -> gim variance -> trnh hin tng overfitting
- Thut ton Random Forest kt hp nhiu cy quyt nh  gim variance

Vn  gim bias:
- Mi m hnh "yu" ch on ng cho mt s tnh hung d liu
- Kt hp nhiu m hnh "yu"  tn dng im mnh ca mi m hnh, khc phc nhng trng hp m tng m hnh on sai

### K thut Bagging

Khc vi 2 thut ton trc, Bagging s dng cng mt thut ton cho tt c m hnh con

Bagging hun luyn c lp cc m hnh con, trn cc tp con ca dataset

in hnh ca k thut ny l Random Forest

u im:
- Hiu qu: chnh xc, tng qut cao
- Tin li: c th thc hin c trn s c trng ln m khng cn phn tch c trng
- Linh hot: dng cho c hi quy v phn lp, c th song ha thut ton
- Bn vng: t b nh hng outlier, t kh nng overfitting

Mt s khuyt im:
- Kh gii thch m hnh
-  phc tp tnh ton cao
- Bias vi d liu khng cn bng
