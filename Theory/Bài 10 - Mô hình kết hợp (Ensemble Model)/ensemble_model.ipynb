## Tại sao cần có Ensemble Model

### Giới thiệu

Mục tiêu máy học: xây dụng mô hình có tính tổng quát hóa cao từ dữ liệu

Có hai cách chính để cải thiện tính tổng quát hóa
- Cải thiện hiệu suất mô hình máy học
- Kết hợp nhiều mô hình và tổ hợp kết quả dự đoán -> Ensemble Learrning

image.png

### Tại sao Ensemble Learning hiệu quả

Vấn đề giảm variance:
- Sử dụng nhiều mô hình có thể trung bình giá trị dự đoán gần với giá trị thực tế -> giảm variance -> tránh hiện tượng overfitting
- Thuật toán Random Forest kết hợp nhiều cây quyết định để giảm variance

Vấn đề giảm bias:
- Mỗi mô hình "yếu" chỉ đoán đúng cho một số tính huống dữ liệu
- Kết hợp nhiều mô hình "yếu" để tận dụng điểm mạnh của mỗi mô hình, khắc phục những trường hợp mà từng mô hình đoán sai

## Kỹ thuật cơ bản

### Kỹ thuật Voting

Thưởng dùng cho bài toán phân loại

Mỗi mô hình như một "cử tri", quyết định cuối cùng thuộc về số đông

image.png

### Kỹ thuật Averaging

Thường dùng cho bài toán hồi quy

Tính trung bình cộng kết quả của từng mô hình để tổng hợp

image.png

### Kỹ thuật Weighted Averaging

Mỗi mô hình có hiệu quả/ trọng số khác nhau nên có trọng số khác nhau

Trọng số có thể được tính từ độ chính xác trên tập train/validation

image.png

## Kỹ thuật nâng cao

### Kỹ thuật Stacking

Sử dụng kết quả dự đoán của tập train làm đặc trưng để huấn luyện mô hình tổng hợp (meta learner)

image.png

### Kỹ thuật Blending

Sử dụng (đặc trưng + kết quả dự đoán của tập validation) làm đặc trưng huấn luyện mô hình tổng hợp

image.png

### Kỹ thuật Bagging

Khác với 2 thuật toán trước, Bagging sử dụng cùng một thuật toán cho tất cả mô hình con

Bagging huấn luyện độc lập các mô hình con, trên các tập con của dataset

Điển hình của kỹ thuật này là Random Forest

Ưu điểm:
- Hiệu quả: chính xác, tổng quát cao
- Tiện lợi: có thể thực hiện được trên số đặc trưng lớn mà không cần phân tích đặc trưng
- Linh hoạt: dùng cho cả hồi quy và phân lớp, có thể song hóa thuật toán
- Bền vững: ít bị ảnh hưởng outlier, ít khả năng overfitting

Một số khuyết điểm:
- Khó giải thích mô hình
- Độ phức tạp tính toán cao
- Bias với dữ liệu không cân bằng

image.png

### Kỹ thuật Boosting

Boosting huấn luyện một cách tuần tự: mô hình sau được train theo kết quả của mô hình trước đó để cố gắng sửa các lỗi sai còn lại

image.png

Gradient Boost: xây dựng các chuỗi cây quyết định liên tiếp, cây sau làm giảm sai số dự đoán của các cây trước

Một số kĩ thuật Boosting phổ biến:
-  Adaboost
- GradientBoost
- XGBoost
- LightGBM
- CatBoost
Đây đều là các thuật toán đạt giải cao trong các cuộc thi Kaggle

## Tổng kết

Ensemble learning là kỹ thuật quan trọng để mô hình có tính tổng quát cao

Bagging và Boosting là hay kỹ thuật nâng cao có tính hiệu quả cao

Trong quá trình sử dụng cần chọn các siêu tham số cho phù hợp bằng phương pháp tinh chỉnh tham số
