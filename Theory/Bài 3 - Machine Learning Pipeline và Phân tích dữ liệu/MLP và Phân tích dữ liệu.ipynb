# Thành phần của Machine Learning Pipeline

## Raw data

- Tổng hợp từ nhiều nguồn: log, database, internet, survey, ...
- Chưa qua kiểm tra, tiền xử lý
- Có thể không đầy đủ, nhiễu, chứa thông tin nhạy cảm

## Data validation

Có vai trò rất quan trọng

- Đảm bảo dữ liệu đầu vào phù hợp và chất lượng
- Kiểm tra: tỉnh nhất quán, đầy đủ, dữ liệu ngoại lệ
- Thường được thực hiện với EDA (Exploratory Data Analysis) - Phân tích dữ liệu

## Data preprocessing

Thực hiện các công việc sau:

- Làm sạch dữ liệu, chuẩn hóa dữ liệu
- Xử lý dữ liệu bị thiếu - Handle mising values
- Xử lý dữ liệu ngoại lệ - Handle outlier values
- Tạo mới đặt trưng - Feature extraction and transformation
- Chọn lựa đặt trưng - Feature selection

## Model training

- Khởi tạo mô hình với các tham số (nếu có)
- Huấn luyện mô hình với dữ liệu (đặc trưng) đã chuẩn bị

## Model evaluation

Tiến hành đánh giá mô hình một cách định lượng

- Dùng bộ dữ liệu thử nghiệm khách quan
- Dùng độ đo (mectric) cụ thể
- Nếu kết quả tốt -> qua bước tiếp theo
- Nếu kết quả không tốt -> quay lại các bước trước

## Model validation

Tiến hành kiểm định mô hình mang tính hệ thống (accept, tốc độ, rò rỉ data, tài nguyên, ...)

- Thường được thực hiện độc lập với đội phát triển mô hình
- Có chính xác trên dữ liệu chưa từng thấy?
- Tốc độ thực thi được đảm bảo?
- Có bị rò rỉ thông tin dữ liệu huấn luyện?

## Model deployment

Triển khai mô hình trên môi trường thực tế

- Đóng gói mô mình
- Chuyển giao mô hình

## Model feedback

Theo dõi phản hồi về mô hình

- Phản hồi chủ quan: nhận xét, đánh giá trực tiếp của người dùng
- Phản hồi khách quan: thông qua các chỉ báo trung gian

# Phân tích dữ liệu - Exploratory Data Analysis

## Tại sao cần Phân tích dữ liệu?

Để hiểu rõ dữ liệu:

- #đặc trưng, #mẫu, loại đặc trưng (số/phân loại), phân bố dữ liệu (min - max)
- Mối quan hệ xu hướng trong dữ liệu

Làm sạch dữ liệu:

- Xác định một số vấn đề: thiếu dữ liệu, trùng lặp  (bias cao), nhiễu, dữ liệu lỗi, hoặc không nhất quán
- Từ đó: xóa/điền khuyết/giữ nguyên

Là tiền đề để:

- Chọn lựa đặc trưng tốt cho bài toán, xấy dựng mô hình
- Có khả năng tạo ra đặc trưng mới tốt hơn
- Chuẩn hóa đặt trưng về dạng phù hợp mô hình máy học

Lường trước một số tình huống:

- Mất cân bằng, thên lệch dữ liệu
- Một số vấn đề khác của dữ liệu: rò rỉ thông tin, vi phạm dữ liệu

Là công cụ truyền thông: trực quan hóa giúp giải thích vấn đề với người không chuyên môn

## Công cụ thực hiện EDA

Công cụ trực quan: histogram, boxplot, ma trận tương quan
Công cụ thống kê:

- Chỉ số thống kê: trung bình, trung vị, phương sai, độ lệch chuẩn
- Phân tích tương quan: đơn biến, đa biến
- Một số công cụ: hồi quy tuyến tính, hệ số tương đồng Pearson, ...


## Các kiểu phân tích dữ liệu

Có 3 kiểu phân tích chính:

- Phân tích đơn biến
- Phân tích hai biến
- Phân tích đa biến

### Phân tích đơn biến

Chỉ quan tâm đến một đặt trưng/cột trong dữ liệu

Phân tích bằng thống kê:

- Xu hướng tập trung (Central Tendency): các tham số ước lượng như trung bình, trung vị, mode
- Khoảng giá trị (Range): khoảng cách giữa giá trị tối đa và tối thiểu trong dữ liệu
- Phương sai và độ lệch chuẩn

Phân tích đơn biến bằng thống kê

- Ví dụ về trung bình, độ lệch chuẩn, tứ phân vị thứ nhất, ba và trung vị của các biến trong dữ liệu

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
df = pd.DataFrame(data=load_iris().data, columns=load_iris().feature_names)
df
```

```python
df.describe()
```

Phân tích đơn biến bằng biểu đồ:

- Biểu đồ Histogram: biểu đồ trong đó tần số của dữ liệu được biểu thị bằng các thanh hình chữ nhật
- Biểu đồ mật độ: một phiên bản liên tục của biểu đồ histogram
- Box-plot: thông tin được thể hiện dưới dạng các hộp (giá trị nhỏ nhất, tứ phân vị thứ nhất (first quartile), trung vị (median). tứ phân vị ba (third quartile). giá trị lớn nhất)

Ví dụ phân tích đơn biến

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x=df['petal length (cm)'][:30])
```

### Phân tích hai biến

Xác định xem có tồn tại mối kiên hệ thống kê giữa hai biến hay không

Có ba loại chính:
- Phân tích dạng số - số (numeric - numeric)
- Phân tích dạng số - phân loại (numeric - category)
- Phân tích dạng phân loại - phân loại (category - category)

Phân tích dạng số - số

- Khi cả hai biển được so sánh đều là dữ liệu số
- Một số phương pháp trực quan có thể sử dụng:


    1. Biểu đồ phân tán (Scatter plot): thể hiện mọi điểm dữ liệu trên bản đồ
    2. Biểu đồ cặp (Pair plot)
    3. Ma trận tương quan (Correlation matrix)



```python
plt.scatter(x=df['sepal length (cm)'],y=df['petal length (cm)']) # Biểu đồ phân tán
plt.title('Scatter plot')
plt.show()
```

```python
sns.pairplot(data=df) # Biểu đồ cặp
plt.title('Pair plot')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True) # Ma trận tương quan
plt.title('Correlation matrix')
plt.show()
```

Phân tích dạng số - phân loại

- Khi một biến kiểu số và biến khác kiểu phân loại
- Một số phương pháp trực quan có thể sử dụng: Multiple density/Histogram,  Box-plot

```python
df = pd.read_csv("/content/drive/MyDrive/Python_ML/Theory/Data/titanic.csv")
df
```

```python
sns.histplot(data=df, x='Age', hue='Survived') # Histogram
plt.title('Histogram')
plt.show()
```

```python
sns.boxplot(data=df, x='Survived', y='Age') # Box-plot
plt.title('Box-plot')
plt.show()
```

Phân tích dữ liệu dạng phân loại - phân loại

- Khi cả hai biến đều có kiểu phân loại
- Một số phương pháp trực quan có thể sử dụng
    
    1. Biểu đồ cột xếp chồng (Stacked Bar Chart)
    2. Biểu đồ cột theo cụm (Clustered Bar Chart)

```python
sns.countplot(x=df['Sex'], hue=df['Survived']) # Biểu đồ cột xếp chồng
plt.title('Stacked Bar Chart')
plt.show()
```

### Phân tích đa biến

- Khi phân tích nhiều hơn 2 biến cùng lúc
- Để trực quan hóa cần sử dụng nhuần nhuyễn các loại biểu đồ tương ứng với các tổ hợp biến dạng số và dạng phân loại

## Phát hiện dữ liệu bị thiếu

- Sử dụng hàm `isnull()` của dataframe
- Có thể loại bỏ các dòng có giá trị NaN hoặc rổng tvoiws hàm *`*drop_nan()*`*

## Phát hiện dữ liệu bị nhiễu

Phương pháp thống kê:

- Phương pháp tính trung bình và độ lệch chuẩn: để xác định các giá trị ngoại lệ (với dữ liệu dạng Gaussian hoặc tương tự Gassian)
- Phương pháp Interquartile Range(IQR): để xác định các giá trị ngoại lệ với dữ liệu phân phối không phải Gaussian
