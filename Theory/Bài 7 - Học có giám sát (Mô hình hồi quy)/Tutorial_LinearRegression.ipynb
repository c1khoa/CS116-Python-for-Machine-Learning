# 1. Generate dữ liệu

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1.0, 10.0, 0.5)
noise = np.random.normal(0.0, 2.0, size=x.shape)
y = 2*x + 5.0 + noise

plt.plot(x, y, 'ro')
plt.show()
```

# 2. Cài đặt mô hình (Build Model)

```python
from sklearn.linear_model import LinearRegression
x = x.reshape(-1, 1)
lr = LinearRegression()
lr.fit(x, y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

## Trực quan hóa mô hình

```python
# Plot các điểm
plt.plot(x[:, 0], y, 'ro')
plt.plot(x[:, 0], lr.predict(x), 'b-')
plt.show()
```

# 3. Đánh giá mô hình

```python
import math
pred = lr.predict(x)
error = math.sqrt(sum((pred - y)**2/len(y)))
error
```

# 4. Generate dữ liệu phi tuyến

## Chưa thêm feature

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10.0, 0.5)
noise = np.random.normal(0.0, 20.0, size=x.shape)
y = 2*x**2 + 5*x + 5.0 + noise

plt.plot(x, y, 'ro')
plt.show()
x = x.reshape(-1, 1)
```

```python
lr = LinearRegression()
lr.fit(x, y)
```

```python
import math
pred = lr.predict(x)
error = math.sqrt(sum((pred - y)**2/len(y)))
error
```

```python
# Plot các điểm
plt.plot(x[:, 0], y, 'ro')
plt.plot(x[:, 0], lr.predict(x), 'b-')
plt.show()
```

## Thêm features

```python
# Thêm cột x^2
x_new = np.concatenate((x, x**2), axis=1)
x_new.shape
```

```python
lr = LinearRegression()
lr.fit(x_new, y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# Plot các điểm
plt.plot(x[:, 0], y, 'ro')
plt.plot(x[:, 0], lr.predict(x_new), 'b-')
plt.show()
```
