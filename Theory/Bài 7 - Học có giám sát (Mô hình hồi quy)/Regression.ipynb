# Học có giám sát - Mô hình hồi quy (Regression)

## Học có giám sát

Học có giám sát (supervised learrning): là một nhánh của máy học, nhằm dự đoán giá trị đầu ra từ một đặt trưng đầu vào dựa trên các dữ liệu huấn luyện trước đó

Dữ liệu huấn luyện bao gồm đặt trưng đầu vào và giá trị đầu ra mong muốn (x, y)

Dữ liệu x -> Mô hình máy học f(x) -> dự doán y_prep ~ y mong muốn

Có hai loại bài toán chính: hồi quy và phân lớp

## Mô hình hồi quy tuyến tính - Linear Regression

Mô hình thực tế: $y = \varepsilon + \beta x$  
Hàm mô hình dự đoán: $\hat{y} = \hat{\beta}x$  

Hàm độ lỗi: $ \text{MSE} = L(\hat{\beta}) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \sum_{i=1}^{n} (y_i - \hat{\beta} x_i)^2 = \| Y - \hat{\beta} X \|^2
$

Huấn luyện mô hình:
- Bằng normal equation: tham số ước lượng: $ \hat{\beta} = (X^TX)^{-1}(X^TY) $
- Bằng thuật toán Gradient descent với công thức cập nhật: $ \hat{\beta} = \hat{\beta} - \alpha \Delta_{\hat{\beta}}L $

Trong sklearn đã cà đặt module: LinearRegression

Ưu điểm:
- Đơn giản, dễ hiểu
- Phù hợp với dữ liệu có mối quan hệ tuyến tính (đồng biến, nghịch biến)

Khuyết điểm:
- Không hiệu quả khi dữ liệu có mối quan hệ phức tạp
- Dễ bị ảnh hưỡng dữ liệu nhiễu



## Bias và Variance

Bias: là sai số giữa trung bình (kỳ vọng) mô hình dự đoán với mô hình thực tế:
$$ bias(\hat{\beta}) = E(\hat{\beta}) - \beta $$
- Bias thấp: thể hiện mô hình học được mối quan hệ của dữ liệu huấn luyện
- Bias cao: thể hiện mô hình không học được mối quan hệ của dữ liệu huấn luyên

Variance: là sai số trung bình của dự đoán so với kỳ vọng (trung bình) mô hình được huấn luyện trên toàn bộ dữ liệu thực: $$ variance(\hat{\beta}) = E[(E(\hat{\beta}) - \beta)^2] $$
- Variance thấp: thể hiện tính tổng quát cao của mô hình, dù huấn luyện trên một tập con vẫn đoán đúng trên tập dữ liệu chưa thấy (tập test)
- Variance cao: thể hiện mô hình không đoán tốt trên dữ liệu chưa gặp

### Overfitting

Nguyên nhân: xảy ra khi mô hình quá phức tạp hoặc dữ liệu không đủ khái quát

Để tránh overfitting:
- Giảm bớt sự phức tạp của mô hình (giảm thâm số)
- Lấy mẫu thêm dữ liệu để bao gồm các tình huống thực tế (tăng dữ liệu)

### Undeffiting

Nguyên nhân: xùy ra khi mô hình quá đơn giản so với tính chất phức tạp của dữ liệu, hoặc dữ liệu không đủ khái quát

Để tránh overfitting:
- Chuyển sang mô hình có khả năng biểu diễn phức tạp hơn (thay mô hình)
- Lấy mẫu thêm dữ liệu để bao gồm các tình huống thực tế (tăng dữ liệu)

## Mô hình Lasso, Ridge và Elastiic Net

### Lasso Regression

Hồi quy tuyển chính + chính quy hóa $L_1$

Có thể dùng để lựa chọn đặc trưng (Feature selection): mô hình cố gắng đưa ra các hệ số về với 0 đối với các đặc trưng không quan trọng
$$
\arg\min_{\hat{\beta}} \left\| Y - \hat{\beta}X \right\|^2 + \lambda \left\| \hat{\beta} \right\|_1
$$
-> Hướng đến lựa chọn đặt trưng quan trọng


### Ridge Regression

Hồi quy tuyến tính + chính quy hóa L2
$$
\arg\min_{\hat{\beta}} \left\| Y - \hat{\beta}X \right\|^2 + \lambda \left\| \hat{\beta} \right\|_2^2
$$
-> Hướng đến khái thác hết các đặc trưng

### Elastic net

Hồi quy tuyến tính + chính quy hóa $ L_1 + L_2 $
$$
\arg\min_{\hat{\beta}} \left\| Y - \hat{\beta}X \right\|^2 + \lambda \left( \frac{1 - \alpha}{2} \left\|\hat{\beta} \right\|_2^2 + \alpha \left\| \hat{\beta} \right\|_1 \right)
$$
-> Cân bằng các tiêu chí của Lasso và Ridge Regression
